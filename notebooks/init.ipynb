```python
%sql
CREATE CATALOG if not exists crypto;
```

```python
%sql
USE CATALOG crypto;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS crypto.bronze
COMMENT 'Bronze layer: raw ingested data';

CREATE SCHEMA IF NOT EXISTS crypto.silver
COMMENT 'Silver layer: cleaned and transformed data';

CREATE SCHEMA IF NOT EXISTS crypto.gold
COMMENT 'Gold layer: business-ready data';
```

```python
%sql
SHOW SCHEMAS;
```

```python
%sql
CREATE VOLUME if not exists crypto.bronze.crypto_raw;
```
